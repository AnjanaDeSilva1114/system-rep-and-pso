```python
import numpy as np
from scipy.signal import lti, step, impulse, freqresp
import matplotlib.pyplot as plt

# Define the system's coefficients
numerator = [13, 9]  # Coefficients of the numerator of the transfer function
denominator = [1, 5, 17, 11]  # Coefficients of the denominator of the transfer function

# Create a transfer function
system = lti(numerator, denominator)

# Step 2: Response to an impulse input
time_impulse, response_impulse = impulse(system)
plt.figure()
plt.plot(time_impulse, response_impulse)
plt.title('Response to Impulse Input')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# Step 3: Response to a step input
time_step, response_step = step(system)
plt.figure()
plt.plot(time_step, response_step)
plt.title('Response to Step Input')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# Step 4: Pole-zero plot
plt.figure()
system_poles, system_zeros = system.poles, system.zeros
plt.scatter(np.real(system_poles), np.imag(system_poles), marker='x', label='Poles')
plt.scatter(np.real(system_zeros), np.imag(system_zeros), marker='o', label='Zeros')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Pole-Zero Plot')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()
plt.show()

# Step 5: Bode plot
freq, response = freqresp(system)
mag = np.abs(response)
phase = np.angle(response, deg=True)  # Convert phase to degrees

plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(freq, 20 * np.log10(mag))
plt.title('Bode Plot - Magnitude')
plt.xlabel('Frequency')
plt.ylabel('Magnitude (dB)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.semilogx(freq, phase)
plt.title('Bode Plot - Phase')
plt.xlabel('Frequency')
plt.ylabel('Phase (degrees)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
